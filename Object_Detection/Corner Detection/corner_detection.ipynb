```python
# Corner:
#     => A corner is a point whose local neighborhood stands in two
#        dominant and different edge direction.
#     => A corner cab be interpreted as the junction of two edges
#        where an edge is a sudden change in image brightness.

# Corner detection algorithms:
#     * Harris corner detection
#     * Shi-Tomasi corner detection
```

```python
# Harris corner detection:
#     => Published in 1988 by Chiris Harris and Mike Stephens.
#     => The basic intuition is that corners can be detected by looking
#        for significant changes in all direction.
```

```python
# Shi-Thomasi corner detection:
#     => Published in 1994 by J. Shi and C. Tomasi in their paper Good Feature to Track.
#     => It made a small modification to the Harrish Corner Detection which ended up with better results.
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../../data/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat_chess)
```

```python
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess = cv2.imread('../../data/real_chessboard.jpg')
real_chess =  cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess)
```

```python
plt.imshow(gray_real_chess,cmap='gray')
```

```python
gray_flat_chess
```

```python
 gray = np.float32(gray_flat_chess)
```

```python
gray
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst
```

```python
plt.imshow(dst)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst > 0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(dst)
```

```python
plt.imshow(dst,cmap='gray')
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)

dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
plt.imshow(dst)
```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(real_chess)
```

```python
plt.imshow(real_chess,cmap='gray')
```
