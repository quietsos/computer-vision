```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
sohan = cv2.imread('sohan.jpg')
sohan = cv2.cvtColor(sohan,cv2.COLOR_BGR2RGB)
plt.imshow(sohan)
```

```python
sohan_face = cv2.imread('sohan_face.jpg')
sohan_face = cv2.cvtColor(sohan_face,cv2.COLOR_BGR2RGB)
plt.imshow(sohan_face)
```

```python
method = cv2.TM_CCOEFF
res = cv2.matchTemplate(sohan,sohan_face,method)
```

```python
plt.imshow(res)
```

```python
min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
```

```python
topLeft = max_loc
```

```python
sohan_face.shape
height,width,channels= sohan_face.shape
```

```python
botRight = (topLeft[0]+width,topLeft[1]+height)
```

```python
cv2.rectangle(sohan,topLeft,botRight,(255,0,0),5)
```

```python
plt.imshow(sohan)
```
