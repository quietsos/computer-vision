```python
# Histogram:
#     A histogram is a visual representation
#     of the distribution of a continous feature.
```

```python
# For image:
#     * We can display the frequency of values for colors.
#     * Each of the three RGB channels has values between 0-255
#     * We can plot these tas 3 histograms on top of each other to se
#          how much of each channel there is.
    
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('../data/horse.jpg') #ORIGINAL BGR OPENCV
show_horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) # CONVERTED TO RGB FOR SHOW

rainbow = cv2.imread('../data/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('../data/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dark_horse)
```

```python
plt.imshow(show_horse)
```

```python
plt.imshow(rainbow)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(blue_bricks)
```

```python
plt.imshow(show_bricks)
```

```python
# Calculate the histogram:
# BGR
blue_hist = cv2.calcHist([blue_bricks],channels=[0],mask = None, histSize=[256],ranges=[0,256])
```

```python
blue_hist.shape
```

```python
plt.plot(blue_hist)
```

```python
# Blue channel hsitogram for dark_horse:

blue_dark_horse = cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(blue_dark_horse)
```

```python
# Rainbow blue channel histogram:
blue_rainbow_channel = cv2.calcHist([rainbow],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(blue_rainbow_channel)
```

```python
#Three color histogram for blue bricks:BGR
img = blue_bricks
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])

plt.title("Histogram for blue bricks:")
```

```python
#Three color histogram for horse:BGR
img = dark_horse
color=('b','g','r')

for i,col in enumerate(color):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim([0,50])
    plt.ylim([0,500000])

plt.title("Histogram for horse:")
```
