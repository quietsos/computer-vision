```python
#Draw circle on a given image

import cv2
import numpy as np


#function
def draw_img(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(read_img,(x,y),50,(0,0,255),2)


#connect open image with callback function
cv2.namedWindow(winname='group_pic')
cv2.setMouseCallback('group_pic',draw_img)


#show image
read_img = cv2.imread('../data/sohan.jpg')
# circle_img = cv2.cvtColor(read_img,cv2.COLOR_BGR2RGB)


while True:
    cv2.imshow('group_pic',read_img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()

```
