```python
import cv2



# Callback  function

def draw_circle(event,x,y,flags,param):
    
    global pt,lefTop_click
    
    if event == cv2.EVENT_LBUTTONDBLCLK:
        pt = (x,y)
        lefTop_click = False
        
    if event == cv2.EVENT_LBUTTONDOWN:
        pt = (x,y)
        lefTop_click = True
        
#         if lefTop_click == True:
#             pt = (0,0)
#             lefTop_click = False
        
#         elif lefTop_click == False:
#             pt = (x,y)
#             lefTop_click = True
        
            



# connection with callback function
cap = cv2.VideoCapture(0)



cv2.namedWindow('Sohan')
cv2.setMouseCallback('Sohan',draw_circle)


# global variables

pt = (0,0)
lefTop_click = False



while True:
    ret,frame = cap.read()
    
    
    if lefTop_click == True:
        cv2.circle(frame,pt,40,(255,0,0),2)
        
    
    cv2.imshow('Sohan',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        

cap.release()
cv2.destroyAllWindows()
```
